Три автора поделили гонорар в отношении 5:6:8. Если бы они его делили в отношении 4:5:7, то один из них
получил бы на 25 руб. больше. Сколько всего было денег?

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

print(solve(Eq(4 * x / 16 - 5 * x / 19, 25))) 
print(solve(Eq(5 * x / 16 - 6 * x / 19, 25)))
print(solve(Eq(7 * x / 16 - 8 * x / 19, 25)))
```
